```python
import json
data=[{
    'name': 'Among Us',
    'price': 10.00,
    'rating': 4.8
},
{
    'name': 'Cyberpunk 2077',
    'price': 29.99,
    'rating': 4.2
}
]
```

```python
json_str = json.dumps(data) # load data above to JSON
```

```python
# convert json back to python data structure
json_data = json.loads(json_str)
print(json_data)
```

Load and convert json, but to a file from HTTP request  of page src code (cookbook)

```python
import requests

data = json.loads(requests.get("https://jsonplaceholder.typicode.com/todos").text)

with open("example.json",'w') as f:
    json.dump(data,f)
try:
    with open("example.json",'r') as f:
        json.load(f)
        # alphabetize keys when printing for readability
except(FileNotFoundError):
    print("No such file found")
```

```python
from pprint import pprint
try:
    with open("example.json",'r') as f:
        # alphabetize keys when printing for readability
        pprint(json.load(f))
except(FileNotFoundError):
    print("No such file found")
```

Dump-load does not return perfect copy:
Ex: tuple data after toJSON and back will return a list

```python
blackjack_hand = (8, "Q")
encoded_hand = json.dumps(blackjack_hand)
decoded_hand = json.loads(encoded_hand)
```
